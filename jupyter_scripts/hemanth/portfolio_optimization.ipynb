```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import yfinance as yf
from pandas_datareader import data as pdr
import itertools
%matplotlib notebook
from pandas.plotting import register_matplotlib_converters
```

```python
def get_stock_data(stock_tickers, start, now):
    stock_df = pd.DataFrame()
    for ticker in stock_tickers:
        print(f"Analysis for {ticker}")
        stock_df[ticker] = pdr.get_data_yahoo(ticker, start, now)['Adj Close']
    return stock_df

def plot_group_grphs(dataframe):
    
    fig, (ax1, ax2) = plt.subplots(2, 1)
    for col in dataframe.columns.values:
        df = pd.DataFrame()
        ax1.plot(dataframe[col], label=col)

assets = ['AAPL', 'GOOG', 'MSFT', 'M', 'DAL', 'MU']
weights = np.array([0.2,0.2,0.2,0.1,0.1,0.2])

start = datetime.now() - timedelta(days=365)
now = datetime.now()
stock_df = get_stock_data(assets, start, now)
stock_df
```

```python
plot_group_grphs(stock_df)
```

```python
returns = stock_df.pct_change()
returns
```

```python
cov_matrix_annaul = returns.cov()*252
cov_matrix_annaul
```

```python
port_variance = np.dot(weights.T, np.dot(cov_matrix_annaul, weights.T))
port_voltality = np.sqrt(port_variance)
port_annual_return = np.sum(returns.mean()*weights)*252
print(port_annual_return, port_voltality, port_variance)
```
